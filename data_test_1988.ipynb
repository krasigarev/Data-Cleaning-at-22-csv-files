```python
import pandas as pd
```

```python
df = pd.read_csv('dataverse_files/1988.csv.bz2', engine='pyarrow')
df.head()
```

```python
def adult(df):
    summary = pd.DataFrame({     
        'dtype': df.dtypes,
        'nulls': df.isnull().sum(),
        'nulls_%': (df.isnull().mean() * 100).round(2),
        'nunique': df.nunique(),
        'sample': df.iloc[0],
    })
    return summary
```

```python
adult(df)
```

```python
# drop columns
col = ['CarrierDelay', 'WeatherDelay', 'NASDelay', 'SecurityDelay',  'LateAircraftDelay', 'TaxiIn', 'TaxiOut', 'AirTime', 'TailNum', 'CancellationCode']
df.drop(columns=col, inplace=True)
```

```python
adult(df)
```

```python
df['Distance'] = df['Distance'].fillna(0)
```

```python
df.head()
```

```python
df['DepTime'] = df['DepTime'].fillna(df['DepTime'].mean())
df['ArrTime'] = df['ArrTime'].fillna(df['ArrTime'].mean())
df['ActualElapsedTime'] = df['ActualElapsedTime'].fillna(df['ActualElapsedTime'].mean())
df['ArrDelay'] = df['ArrDelay'].fillna(df['ArrDelay'].mean())
df['DepDelay'] = df['DepDelay'].fillna(df['DepDelay'].mean())
```
